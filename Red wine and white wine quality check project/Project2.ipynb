# Q14.Create a new column as the acidity level. Divide the existing data into 5 groups based on the acidity level. The levels are to be named ['High', 'Moderately_High', 'Medium', 'Low']. Acidity may be obtained from the PH.

```python
# Create a new column 'acidity_level'
df_combined['acidity_level'] = None

# Divide the data into 5 groups based on pH value
for i, row in df_combined.iterrows():
    if row['pH'] >= 3.0 and row['pH'] < 3.3:
        df_combined.at[i, 'acidity_level'] = 'High'
    elif row['pH'] >= 3.3 and row['pH'] < 3.5:
        df_combined.at[i, 'acidity_level'] = 'Moderately_High'
    elif row['pH'] >= 3.5 and row['pH'] < 4.0:
        df_combined.at[i, 'acidity_level'] = 'Medium'
    else:
        df_combined.at[i, 'acidity_level'] = 'Low'
```

# Q21.Is the density of red wine and white wine similar?

```python
from scipy.stats import ttest_ind

red_density = df_combined[df_combined['color'] == 'red']['density']
white_density = df_combined[df_combined['color'] == 'white']['density']

t, p = ttest_ind(red_density, white_density, equal_var=False)

if p < 0.05:
    print("The means of the density of red and white wine are significantly different.")
else:
    print("The means of the density of red and white wine are not significantly different.")
```

# Q22.Is the confidence interval for the differences of the proportion of white wine with rating 3 and the proportion of red wine with rating 3.

```python
white_wines = df_combined[df_combined['color'] == 'white']
num_white_wines = len(white_wines)
```

```python
red_wines = df_combined[df_combined['color'] == 'red']
num_red_wines = len(red_wines)
```

```python
print(num_white_wines)
print(num_red_wines)
```

```python
red_wines_rating_3 = df_combined[(df_combined['color'] == 'red') & (df_combined['quality'] == 3)]
num_red_wines_rating_3 = len(red_wines_rating_3)
```

```python
num_white_wines_rating_3 = df_combined[(df_combined['color'] == 'white') & (df_combined['quality'] == 3)].shape[0]
```

```python
print(num_red_wines_rating_3)
print(num_white_wines_rating_3)
```

```python
from scipy.stats import norm
import numpy as np

# Counts of white wines with rating 3
n1 = num_white_wines_rating_3
p1 = n1 / num_white_wines

# Counts of red wines with rating 3
n2 = num_red_wines_rating_3
p2 = n2 / num_red_wines

# Calculate standard error of the difference in proportions
SE = np.sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)

# Desired level of confidence (e.g., 95%)
confidence = 0.95
z = norm.ppf((1 + confidence) / 2)

# Calculate the confidence interval
CI = (p1 - p2) + np.array([-1, 1]) * z * SE

print("Confidence interval:", CI)
```
